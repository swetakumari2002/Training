```python
def whisper(text):
	return text.lower()

def greet(func):
	greeting = func("""Hi, I am created by a function
					passed as an argument.""")
	print (greeting)


greet(shout)
greet(whisper)
```

```python
def create_adder(x):
	def adder(y):
		return x**y

	return adder

add_15 = create_adder(25)

print (add_15(20))
```
